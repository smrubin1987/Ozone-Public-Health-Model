Findings from EDA

Distribution of daily 8-hour O3  max concentrations per county differs. Los Angeles county appears to have a higher distribution of max O3 concentrations, with a significant portion over both 0.06ppm and 0.07ppm. Riverside county is close behind. Both counties are in California.


The distplots show that there are days where ground-level O3 does measure well above 0.06ppm (where it may start causing health effects), above 0.07ppm (the EPA threshold for attainment), and even has high as 0.10ppm (which is dangerously harmful). 

Is there an association with county-level total population population?
