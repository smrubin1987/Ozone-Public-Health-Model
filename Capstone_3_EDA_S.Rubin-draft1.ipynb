```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
Ozone_df = pd.read_csv("AllCountiesMeanOzone.csv")
del Ozone_df['Unnamed: 0']
Ozone_df.head()
```

```python
#Summary stats of one-hour ozone measurements by county by day for 2019
Ozone_df.head()
Ozone_df.describe()

#Orange County had the highest max 1-hour ozone at 0.10067; well above the CA-EPA ambient air quality standard of 0.09ppm for 1-hour average
#Harris County had the lowest mean 1-hour ozone at 0.02372 ppm
#min records of ozone are likely attributable to errors or mis-measurments
#highest median 1-hour ozone was Riverside at 0.03850, lowest median (50%) was Harris County at 0.02187
```

```python
#max daily ppm by county
Max_Maricopa = pd.DataFrame(pd.Series(Ozone_df.groupby('date_local').Ozone_Maricopa.max()))
Max_Maricopa['County'] = 'Maricopa'
Max_Maricopa = Max_Maricopa.rename(columns={'Ozone_Maricopa': 'Ozone Max (ppm)'})

Max_LA = pd.DataFrame(pd.Series(Ozone_df.groupby('date_local').Ozone_LosAngeles.max()))
Max_LA['County'] = 'LosAngeles'
Max_LA = Max_LA.rename(columns={'Ozone_LosAngeles': 'Ozone Max (ppm)'})

Max_Cook = pd.DataFrame(pd.Series(Ozone_df.groupby('date_local').Ozone_Cook.max()))
Max_Cook['County'] = 'Cook'
Max_Cook = Max_Cook.rename(columns={'Ozone_Cook': 'Ozone Max (ppm)'})

Max_SanDiego = pd.DataFrame(pd.Series(Ozone_df.groupby('date_local').Ozone_SanDiego.max()))
Max_SanDiego['County'] = 'SanDiego'
Max_SanDiego = Max_SanDiego.rename(columns={'Ozone_SanDiego': 'Ozone Max (ppm)'})

Max_Harris = pd.DataFrame(pd.Series(Ozone_df.groupby('date_local').Ozone_Harris.max()))
Max_Harris['County'] = 'Harris'
Max_Harris = Max_Harris.rename(columns={'Ozone_Harris': 'Ozone Max (ppm)'})

Max_Orange = pd.DataFrame(pd.Series(Ozone_df.groupby('date_local').Ozone_Orange.max()))
Max_Orange['County'] = 'Orange'
Max_Orange = Max_Orange.rename(columns={'Ozone_Orange': 'Ozone Max (ppm)'})

Max_MiamiDade = pd.DataFrame(pd.Series(Ozone_df.groupby('date_local').Ozone_MiamiDade.max()))
Max_MiamiDade['County'] = 'MiamiDade'
Max_MiamiDade = Max_MiamiDade.rename(columns={'Ozone_MiamiDade': 'Ozone Max (ppm)'})

Max_Dallas = pd.DataFrame(pd.Series(Ozone_df.groupby('date_local').Ozone_Dallas.max()))
Max_Dallas['County'] = 'Dallas'
Max_Dallas = Max_Dallas.rename(columns={'Ozone_Dallas': 'Ozone Max (ppm)'})

Max_Riverside = pd.DataFrame(pd.Series(Ozone_df.groupby('date_local').Ozone_Riverside.max()))
Max_Riverside['County'] = 'Riverside'
Max_Riverside = Max_Riverside.rename(columns={'Ozone_Riverside': 'Ozone Max (ppm)'})

Max_Queens = pd.DataFrame(pd.Series(Ozone_df.groupby('date_local').Ozone_Queens.max()))
Max_Queens['County'] = 'Queens'
Max_Queens = Max_Queens.rename(columns={'Ozone_Queens': 'Ozone Max (ppm)'})

#concat each county to a new, single dataframe for max daily ozone values
all_df_max = pd.concat([Max_Maricopa, Max_LA, Max_MiamiDade, Max_Dallas, Max_SanDiego, Max_Cook, Max_Harris, Max_Riverside, Max_Queens, Max_Orange])
all_df_max.shape
all_df_max.head()
```

```python
#mean daily ppm by county
Ozone_df_mean_Maricopa = pd.Series(Ozone_df.groupby('date_local').Ozone_Maricopa.mean())
Maricopa1_df = Ozone_df_mean_Maricopa.reset_index()
Maricopa1_df['County'] = 'Maricopa'
Maricopa1_df = Maricopa1_df.rename(columns={'Ozone_Maricopa': 'Ozone Mean (ppm)'})

Ozone_df_mean_LosAngeles = pd.Series(Ozone_df.groupby('date_local').Ozone_LosAngeles.mean())
LosAngeles1_df = Ozone_df_mean_LosAngeles.reset_index()
LosAngeles1_df['County'] = 'LosAngeles'
LosAngeles1_df = LosAngeles1_df.rename(columns={'Ozone_LosAngeles': 'Ozone Mean (ppm)'})

Ozone_df_mean_Cook = pd.Series(Ozone_df.groupby('date_local').Ozone_Cook.mean())
Cook1_df = Ozone_df_mean_Cook.reset_index()
Cook1_df['County'] = 'Cook'
Cook1_df = Cook1_df.rename(columns={'Ozone_Cook': 'Ozone Mean (ppm)'}) 

Ozone_df_mean_SanDiego = pd.Series(Ozone_df.groupby('date_local').Ozone_SanDiego.mean())
SanDiego1_df = Ozone_df_mean_SanDiego.reset_index()
SanDiego1_df['County'] = 'SanDiego'
SanDiego1_df = SanDiego1_df.rename(columns={'Ozone_SanDiego': 'Ozone Mean (ppm)'}) 

Ozone_df_mean_Harris = pd.Series(Ozone_df.groupby('date_local').Ozone_Harris.mean())
Harris1_df = Ozone_df_mean_Harris.reset_index()
Harris1_df['County'] = 'Harris'
Harris1_df = Harris1_df.rename(columns={'Ozone_Harris': 'Ozone Mean (ppm)'}) 

Ozone_df_mean_Orange = pd.Series(Ozone_df.groupby('date_local').Ozone_Orange.mean())
Orange1_df = Ozone_df_mean_Orange.reset_index()
Orange1_df['County'] = 'Orange'
Orange1_df = Orange1_df.rename(columns={'Ozone_Orange': 'Ozone Mean (ppm)'}) 

Ozone_df_mean_MiamiDade = pd.Series(Ozone_df.groupby('date_local').Ozone_MiamiDade.mean())
MiamiDade1_df = Ozone_df_mean_MiamiDade.reset_index()
MiamiDade1_df['County'] = 'MiamiDade'
MiamiDade1_df = MiamiDade1_df.rename(columns={'Ozone_MiamiDade': 'Ozone Mean (ppm)'}) 

Ozone_df_mean_Dallas = pd.Series(Ozone_df.groupby('date_local').Ozone_Dallas.mean())
Dallas1_df = Ozone_df_mean_Dallas.reset_index()
Dallas1_df['County'] = 'Dallas'
Dallas1_df = Dallas1_df.rename(columns={'Ozone_Dallas': 'Ozone Mean (ppm)'}) 

Ozone_df_mean_Riverside = pd.Series(Ozone_df.groupby('date_local').Ozone_Riverside.mean())
Riverside1_df = Ozone_df_mean_Riverside.reset_index()
Riverside1_df['County'] = 'Riverside'
Riverside1_df = Riverside1_df.rename(columns={'Ozone_Riverside': 'Ozone Mean (ppm)'}) 

Ozone_df_mean_Queens = pd.Series(Ozone_df.groupby('date_local').Ozone_Queens.mean())
Queens1_df = Ozone_df_mean_Queens.reset_index()
Queens1_df['County'] = 'Queens'
Queens1_df = Queens1_df.rename(columns={'Ozone_Queens': 'Ozone Mean (ppm)'}) 

#concat county-specifc daily mean ozone to a new dataframe
all_df_mean = pd.concat([Maricopa1_df, LosAngeles1_df, MiamiDade1_df,Dallas1_df, SanDiego1_df, Cook1_df, 
                    Harris1_df, Riverside1_df,Queens1_df, Orange1_df])
all_df_mean.shape
all_df_mean.head()
```

```python
#reset_index of both dataframes, concat into a new, combined dataframe with 'date_local', 'county', 'Ozone Mean (ppm)', and 'Ozone Max (ppm)'
all_df_mean.reset_index(drop=True, inplace=True)
all_df_max.reset_index(drop=True, inplace=True)
Ozone_DF1 = pd.concat([all_df_mean, all_df_max], axis=1)
Ozone_DF1.head()
Ozone_DF1.shape  #right shape, need to drop a few redundant columns

Ozone_df_final = Ozone_DF1.loc[:,~Ozone_DF1.columns.duplicated()]
Ozone_df_final.head() #yes!
```

```python
#swarmplot, violin plot of daily mean ozone by county
sns.swarmplot(x='County', y='Ozone Mean (ppm)', data=Ozone_df_final, size=2)    #y='Ozone (ppm)', x='County',
sns.violinplot(x='County', y='Ozone Mean (ppm)', data=Ozone_df_final)
plt.title('County Daily Mean Ozone (ppm)')
plt.xticks(rotation=45)
```

```python
#swarmplot, violin plot of daily mean ozone by county
sns.swarmplot(x='County', y='Ozone Max (ppm)', data=Ozone_df_final, size=2)    #y='Ozone (ppm)', x='County',
sns.violinplot(x='County', y='Ozone Max (ppm)', data=Ozone_df_final)
plt.title('County Daily Max Ozone (ppm)')
plt.xticks(rotation=45)
```

```python
sns.boxplot(x='County', y='Ozone Mean (ppm)', data=all_df_mean)
plt.title('County Daily Mean Ozone (ppm)')
plt.xticks(rotation=45)
```
